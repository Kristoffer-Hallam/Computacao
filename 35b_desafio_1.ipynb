# Desafio

Escreva um programa em Python que verifique se uma senha fornecida como uma string cumpre os seguintes critérios de segurança:

- Tem pelo menos 8 caracteres.
- Contém pelo menos uma letra maiúscula.
- Contém pelo menos uma letra minúscula.
- Contém pelo menos um número.
- Contém pelo menos um caractere especial (considere caracteres especiais como `!@#$%^&*()-_=+`).

Se a senha atender a todos os critérios, o programa deve imprimir `"Senha válida"`. Caso contrário, deve imprimir `"Senha inválida"`.

Dica: pesquise pelos métodos  `isupper`, `islower`, `isdigit` de strings.

```python
senha = "Senha@123"

# Inicializa os critérios como False
tem_oito_ou_mais_caracteres = len(senha) >= 8
tem_letra_maiuscula = False
tem_letra_minuscula = False
tem_numero = False
tem_caractere_especial = False

# Lista de caracteres especiais para verificação
caracteres_especiais = "!@#$%^&*()-_=+"

# Loop para verificar cada caractere
for c in senha:
    if c.isupper():
        tem_letra_maiuscula = True
    elif c.islower():
        tem_letra_minuscula = True
    elif c.isdigit():
        tem_numero = True
    elif c in caracteres_especiais:
        tem_caractere_especial = True

# Verificação da senha
if (
    tem_oito_ou_mais_caracteres
    and tem_letra_maiuscula
    and tem_letra_minuscula
    and tem_numero
    and tem_caractere_especial
):
    print("Senha válida")
else:
    print("Senha inválida")
```

```python
# forma mais elaborada com uso da função built-in any
senha = "Senha@123"

caracteres_especiais = "!@#$%^&*()-_=+"

# Critérios
tem_oito_ou_mais_caracteres = len(senha) >= 8
tem_letra_maiuscula = any([c.isupper() for c in senha])
tem_letra_minuscula = any([c.islower() for c in senha])
tem_numero = any([c.isdigit() for c in senha])
tem_caractere_especial = any([c in "!@#$%^&*()-_=+" for c in senha])

# Verificação da senha
if all(
    [
        tem_oito_ou_mais_caracteres,
        tem_letra_maiuscula,
        tem_letra_minuscula,
        tem_numero,
        tem_caractere_especial,
    ]
):
    print("Senha válida")
else:
    print("Senha inválida")
```
