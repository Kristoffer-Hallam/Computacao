Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
# Instalar
# pyautogui -> automaçoes do mouse e teclado
# pandas -> manipulaçao de dados
# openpyxl -> ler arquivos excel
# pyperclip -> copiar e colar textos


# Passo a passo do desafio
## Passo 1: Entrar no sistema da empresa (link do drive)
### Manualmente
### abri o Chrome
### digitei o no link do sistema
### apertei enter e esperei carregar a página

# pyautogui.click -> clicar com o mouse
# pyautogui.write -> escrever com o teclado
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> apertar uma combinação de teclas
```

```python
import pyautogui
import pyperclip
import pandas as pd
import time
from pynput.mouse import Controller, Button
```

```python

def open_safari():
    pyautogui.click(x=165, y=735)  # Clica no Dock sobre o ícone do Launchpad
    pyautogui.click(x=339, y=108)  # Clica no ícone do Safari
    # pyautogui.click(x=1307, y=56)  # Clica no ícone de nova aba
    time.sleep(3)
    pyautogui.hotkey('command', 't') # Cria nova aba
    
def pega_coords():
    time.sleep(3)
    print(pyautogui.position())
```

```python
pyautogui.PAUSE = 1  # Pausa de 1 segundo após cada comando do pyautogui
mouse = Controller()

# pyautogui.click(x=165, y=735)  # Clica no Dock sobre o ícone do Launchpad
# pyautogui.click(x=339, y=108)  # Clica no ícone do Safari
# # pyautogui.click(x=1307, y=56)  # Clica no ícone de nova aba
# time.sleep(3)
# pyautogui.hotkey('command', 't') # Cria nova aba
open_safari()
# time.sleep(3)
link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga'
pyautogui.write(link) # Escreve texto na barra de endereços
pyautogui.press('enter')
time.sleep(3)  # Espera 10 segundos para a página carregar
```

```python

## Passo 2: Navegar no sistema para encontrar a base de dados
## Clicar na pasta Exportar
pyautogui.click(x=314, y=312)  # Clica duas vezes na pasta Exportar
mouse.click(Button.left, 2)
time.sleep(3)

## Passo 3: Baixar o arquivo da base de dados (formato .xlsx)
pyautogui.click(x=314, y=312, button='right')  # Clica com o botão direito no arquivo
pyautogui.click(x=362, y=367) # Clica na opção Fazer download
time.sleep(8)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4: Calcular indicadores de desempenho (faturamento e quantidade de produtos vendidos) por loja
## abrir a base de dados
path_to_file = r'/Users/krones/Downloads/Vendas - Dez.xlsx'
tabela = pd.read_excel(path_to_file, engine='openpyxl')

## ver as informações da base de dados
display(tabela)

## somar o faturamento de todos os produtos - somar a coluna Valor Final
faturamento = tabela['Valor Final'].sum()
print(f'O faturamento total foi de R$ {faturamento:,.2f}')

## somar a quantidade de todos os produtos vendidos - somar a coluna Quantidade
quantidade = tabela['Quantidade'].sum()
print(f'A quantidade total de produtos vendidos foi de {quantidade:,} unidades')
```

### Vamos agora enviar um e-mail pelo gmail

```python
## Passo 5: Enviar informações por e-mail para a diretoria
# open_safari()
pyautogui.hotkey('command', 't') # Cria nova aba
pyautogui.write('https://mail.google.com/')
pyautogui.press('enter')
time.sleep(5)

## TALVEZ SEJA NECESSÁRIO FAZER LOGIN MANUALMENTE AQUI
pyautogui.click(x=44, y=208)  # Clica no botão caixa de entrada
pyautogui.click(x=174, y=241)  # Clica no botão Escrever
# mouse.click(Button.left, 1)

#para quem será o email
pyautogui.write('kristoffer.hallam+automacao@gmail.com')
pyautogui.press('tab')  # seleciona o email
pyautogui.press('tab')  # ir para o campo do assunto

# qual o assunto
pyperclip.copy('Relatório de Vendas')
pyautogui.hotkey('command', 'v')
pyautogui.press('tab')  # ir para o corpo do email

# corpo do email
texto = f"""
Prezados,

Segue o relatório de vendas do mês de Dezembro.

Faturamento total: R$ {faturamento:,.2f}
Quantidade total de produtos vendidos: {quantidade:,} unidades

Qualquer dúvida, estou à disposição.

Atenciosamente,
Kristoffer Hallam
"""
pyperclip.copy(texto)
pyautogui.hotkey('command', 'v')

# enviar email
pyautogui.click(x=782, y=680)  # Clica no corpo do email
```
