# Exercícios

Classificar as variáveis a seguir segundo as seguintes escalas: nominal, ordinal, binária, discreta ou contínua.

```
a. Faturamento da empresa.                                  -> quantitativa e contínua
b. Ranking de desempenho: bom, médio e ruim.                -> qualitativa e ordinal
e. Tempo de processamento de uma peça.                      -> quantitativa e contínua
d. Número de carros vendidos.                               -> quantitativa e discreta
e. Distância percorrida em km.                              -> quantitativa e contínua
f. Municípios da Grande São Paulo.                          -> qualitativa e nominal
g. Faixa de renda.                                          -> quantitativa e contínua X qualitativa e nominal
h. Notas de um aluno: A, B, C, D, ou E.                     -> quantitativa e contínua X qualitativa e nominal
i. Horas trabalhadas.                                       -> quantitativa e discreta X contínua
j. Região: Norte, Nordeste, Centro-Oeste, Sul e Sudeste.    -> qualitativa e nominal
k. Tamanho da organização: pequeno, médio e grande porte.   -> qualitativa e ordinal
l. Número de quartos em um hotel.                           -> quantitativa e discreta
m. Classificação de risco: elevado, médio, especulativo, substancial, em moratória. -> qualitativa e ordinal
n. Casado: sim ou não.                                      -> qualitativa e binária
```
