Construa um gráfico de linhas com as seguintes características:

- Tamanho 14x6
- Linha com espessura 3
- Palheta de cores Set3
- Destaque o impacto da pandemia (2020 e 2021) com um vspan
- Destaque o impacto da pandemia com um texto
- Destaque os pontos de início e fim da série com um scatter plot com as quantidades de refeições servidas
- Remova os eixos y
- Remova os ticks do eixo x
- Remova as bordas do gráfico
- Adicione um título
- A figura deve ser salva numa pasta chamada `imagens` que deve estar no mesmo diretório do Jupyter Notebook



```python
# listas de anos e refeições
anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389,
             232797, 110023, 160350, 222789]
```

```python
import os
import matplotlib.pyplot as plt
from cycler import cycler
from altera_paleta_cores import altera

# paleta de cores
# definindo o ciclo de cores
cores = plt.get_cmap('Set3').colors
ciclo_cores = cycler(color=cores)
plt.rc('axes', prop_cycle=ciclo_cores)

# Tamanho da figura e plotagem da linha
plt.figure(figsize=(14, 6))
plt.plot(anos, refeicoes, linewidth=3, zorder=1)

# Destaque período crítico
plt.axvspan(2020, 2021, color='gray', alpha=0.3, label='Período Crítico', zorder=0)
plt.text(2020.5, 200000, 'Pandemia COVID-19', fontsize=12, color='black', weight='bold',rotation=90,
         verticalalignment='center', horizontalalignment='center')

# Colocação de pontos com scatterplot
for posicao in (-1,0):
    plt.scatter(anos[posicao], refeicoes[posicao], color=cores[3], s=100,
                label='Refeições Servidas', zorder=2)
    plt.annotate(f"{refeicoes[posicao]:,}", 
                 xy=(anos[posicao], refeicoes[posicao]),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center',
                 fontsize=14,
                 color=cores[3],
                 weight='bold')

# Removendo caixa e eixos
plt.box(False)
plt.tick_params(axis='x', length=0)


plt.title('Número de Refeições Servidas por Ano', fontsize=16, loc='left', weight='bold', color='gray')
# plt.xlabel('Ano', fontsize=14)
plt.ylabel('Número de Refeições', fontsize=14)
# plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(anos)
plt.tight_layout()
os.makedirs('imagens', exist_ok=True)
plt.savefig('imagens/grafico_barras.png', dpi=300, bbox_inches='tight')
plt.show()
```
