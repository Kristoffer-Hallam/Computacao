# Desafio

Escreva um programa em Python que verifique se uma senha fornecida como uma string cumpre os seguintes critérios de segurança:

- Tem pelo menos 8 caracteres.
- Contém pelo menos uma letra maiúscula.
- Contém pelo menos uma letra minúscula.
- Contém pelo menos um número.
- Contém pelo menos um caractere especial (considere caracteres especiais como `!@#$%^&*()-_=+`).

Se a senha atender a todos os critérios, o programa deve imprimir `"Senha válida"`. Caso contrário, deve imprimir `"Senha inválida"`.

Dica: pesquise pelos métodos  `isupper`, `islower`, `isdigit` de strings.

```python
senha = "Senha@123"

carac_especiais = "!@#$%^&*()-_=+"

tamanho_minimo = len(senha) >= 8
print("A senha possui pelo menos 8 caracteres, atendendo ao tamanho mínimo.")

letra_maiuscula = False
letra_minuscula = False
contem_digito = False
contem_carac_especial = False


for char in senha:
    if char.isupper():
        letra_maiuscula = True
        print("A senha possui pelo menos 1 letra maiúscula.")
        break
    elif char.islower():
        letra_minuscula = True
        print("A senha possui pelo menos 1 letra minúscula.")
        break
    
    elif char.isdigit():
        contem_digito = True
        print("A senha possui pelo menos 1 dígito.")
        break

    elif char in carac_especiais:
        contem_carac_especial = True
        print("A senha possui pelo menos 1 caracter especial.")
        break

if (tamanho_minimo and letra_maiuscula and letra_minuscula and
    contem_digito and contem_carac_especial):
    print("Senha válida.")
else:
    print("Senha inválida.")
```

```python
# Outra abordagem
senha = "Senha@123"

carac_especiais = "!@#$%^&*()-_=+"

tamanho_minimo = len(senha) >= 8

letra_maiuscula = any([char.isupper() for char in senha])
letra_minuscula = any([char.islower() for char in senha])
contem_digito = any([char.isdigit() for char in senha])
contem_carac_especial = any([char in carac_especiais for char in senha])

if all([tamanho_minimo, letra_maiuscula, letra_minuscula,
       contem_digito, contem_carac_especial]):
    print("Senha válida.")
else:
    print("Senha inválida.")
```
